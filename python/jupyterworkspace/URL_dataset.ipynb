```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
```

```python
path = '/home/sejal/Documents/datascience/dataset/data/url-shortner/usagov_bitly_data2012-03-16-1331923249.txt'
```

```python
pd.read_json(path)
```

```python
import json
```

```python
file_obj = open(path, mode='r')
data = [json.loads(line) for line in file_obj]
df = DataFrame(data)
```

```python
#top-10 timezones
raw_data = df['tz']

raw_data.fillna('Unknown')
raw_data[raw_data == ''] = 'Unknown'
final = raw_data.value_counts()[:5]
print final
final.plot(kind = 'pie', shadow=True)
plt.show()
```

```python
#Get count of recoerding from windowns and non-windows 
categories = np.where(df['a'].str.contains('Windows'),'Windows','Non-Windows')
s = Series(categories)
s.value_counts()
```

```python
# get TZ wise windows and non windows users
pdf = df[['tz','categories']]

cdf = pdf.fillna('Missing')
cdf[cdf == ' '] = 'Missing'

gb = cdf.groupby(['tz','categories']) 
df3 = gb['categories'].count()

print df3.unstack()

#Same can be done as pivot
pivot = cdf.pivot_table('tz', aggfunc=np.count_nonzero, index='tz', columns='categories')
pivot
```
