```python
def featureEngg(df) :
    #add trend
    df["Trend"] = [i for i in range(len(df))]
    df["open_lag10"] = df.open.shift(10)
    
    #Adds 10 mins lag feature for close
    df["close_lag10"] = df.close.shift(10)
    
    df.drop(df.head(10).index, inplace=True)
```

```python
def getY(nifty):
    #Generate y
    nifty_Y = pd.DataFrame()
    nifty["y"] = nifty.close.shift(-10)
    nifty.y = np.where( nifty.y > nifty.close,1,0)
    
    nifty_Y["y"] = nifty.y
    nifty.drop("y",axis=1,inplace=True)
    
    return nifty_Y
    
```
