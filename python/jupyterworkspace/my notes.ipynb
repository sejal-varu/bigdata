```python
path = "/home/sejal/Documents/datascience/assignments/data/ipl_matches_small.csv"
ipl_matches_array =np.genfromtxt(path, dtype="|S50", skip_header=1, delimiter=",")
```

```python
def get_total_deliveries_played(batsman):
	batsman_col = ipl_matches_array[:,13]
	return len(batsman_col[batsman_col == batsman])


get_total_deliveries_played("SR Tendulkar")
```

```python
def get_wicket_delivery_numbers_array(player):
    played_out = ipl_matches_array[:,[11,20]]
    return played_out[played_out[:,1] == player][:,0]
    
get_wicket_delivery_numbers_array("ST Jayasuriya")
```

```python
def get_toss_win_count(team):
    toss_winner = ipl_matches_array[:,[0,5]]    
    return len(np.unique(toss_winner[toss_winner[:,1] == team][:,0]))
    
get_toss_win_count("Mumbai Indians")
```

```python

def get_all_sixes_filter():
    return ipl_matches_array[:,16].astype(np.int16) == 6

get_all_sixes_filter()
```
