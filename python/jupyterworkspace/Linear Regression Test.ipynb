```python
import pandas as pd
import numpy as np
from pandas import Series,DataFrame 
```

```python
path = '/home/sejal/Documents/datascience/dataset/data/Concrete_Data.xls'

data = pd.read_excel(path)
data.head()
```

```python
data.shape
```

```python
X = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
y.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y_pred = lr.predict(X)
```

```python
pd.DataFrame({"y_act" : y,"y_pred" : y_pred})
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y , y_pred)
```
