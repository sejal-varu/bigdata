```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#y= 5x + 7x square + rand
rng = np.random.RandomState(1)
#a = rng.rand(100)
#b =5 * a + 7 * a * a + 50*rng.randn(100)

a = np.random.randint(100, size = 100, dtype = 'int')
b =5 * a + 7 * a * a + 50*np.random.randint(100, size = 100, dtype = 'int')

#a = np.random
#b[20] = 50
plt.scatter(a, b)
plt.xlabel("Random Values of x")
plt.ylabel("Random Values of Y")
plt.title("Random Plot w/ Outlier")
plt.show()



X_test = np.random.randint(100, size =100)




from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(a.reshape(-1,1), b)
y_pred = lr.predict(X_test.reshape(-1,1))
```

```python
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree =2)
pf.fit_transform(a.reshape(-1,1))


plt.scatter(a,b)
plt.scatter(X_test, y_pred)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(pf.fit_transform(a.reshape(-1,1)),b)
y_pred = lr.predict(pf.transform(X_test.reshape(-1,1)))
```

```python
plt.scatter(a,b)
plt.scatter(X_test, y_pred)
plt.show()
```
