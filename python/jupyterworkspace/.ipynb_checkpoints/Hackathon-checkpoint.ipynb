```python
def get_numeric_features(df):
    numeric_feature = [a for a in range(len(df.dtypes)) if df.dtypes[a] in ['int64','float64']]
    return df.columns[numeric_feature]
```

```python
def get_categorical_features(df):
    numeric_feature = [a for a in range(len(df.dtypes)) if df.dtypes[a] in ['int8','int64','float64']]
    return df.columns.difference(df.columns[numeric_feature])
```

```python
y = df.y
X = df.drop('y', axis=1)
ss = StandardScaler()
```

```python
X.loc[:, X.columns] = ss.fit_transform(X.loc[:, X.columns])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=9)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict_proba(X_test)[:,1]
```

```python
metrics.roc_auc_score(y_test, y_pred)
```
