```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#y= 5x + 6x square + Sin(x)
rng = np.random.RandomState(1)

X = np.random.randint(100, size = 100, dtype = 'int')
y =5 * X + 7 * X * X + np.sin(X) +  300*np.random.randint(100, size = 100, dtype = 'int')

plt.scatter(X, y)
plt.xlabel("Random Values of X")
plt.ylabel("Random Values of Y")
plt.title("Random Plot w/ Outlier")
plt.show()
```

```python
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
```

```python
lr = LinearRegression()
lasso = Lasso()
ridge = Ridge()

pf = PolynomialFeatures(degree =15)
pf.fit_transform(X.reshape(-1,1))

X_test = np.random.randint(100, size =100)

lr.fit(pf.fit_transform(X.reshape(-1,1)),y)
lasso.fit(pf.fit_transform(X.reshape(-1,1)),y)
ridge.fit(pf.fit_transform(X.reshape(-1,1)),y)

y_pred_lr = lr.predict(pf.transform(X_test.reshape(-1,1)))
y_pred_lasso = lasso.predict(pf.transform(X_test.reshape(-1,1)))
y_pred_ridge = ridge.predict(pf.transform(X_test.reshape(-1,1)))


plt.scatter(X,y)
#plt.scatter(X_test, y_pred_lr)
#plt.scatter(X_test, y_pred_lasso)
plt.scatter(X_test, y_pred_ridge)

plt.show()
```
