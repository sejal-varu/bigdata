```python
#lets first read all files in directory
dir_path = '/home/sejal/Documents/datascience/dataset/data/weather_data/weather_data_set_*'
#glob(dir_path)
'''
l1 = []
for ele in glob(dir_path):
    l1.append(pd.read_csv(ele, delimiter='|'))
'''
files = glob(dir_path)
dfs = [pd.read_csv(file, delimiter='|', names=['station','year','datemonth','hm','temp']) for file in files]
#print dfs

df = pd.concat(dfs, ignore_index=True) #concat all dataframes, without ignore_index the indexes are repeated to we need ignore _index as True
#print df

cdf = df.dropna() #cleaned data frame
#cdf

pcdf = cdf[['year','temp']] #projected cleaned data frame
pcdf
```

```python
#pcdf['temp'].max() #gives maximum temp of all years

gdf = pcdf.groupby('year') # this isgrouped data frame, groups data by year
#print type(gdf)
max_temps = gdf['temp'].max() #Now we get max temp on yearly basis
#print max_temps
max_temps.plot(kind='bar')
plt.show()
```

```python
#load data from other file (which is in diffrent format) into this dataframe


#1st load other data from file
data2_path = '/home/sejal/Documents/datascience/dataset/data/weather_data_company_xyz/weather_data_set'
df2 = pd.read_csv(data2_path, names=['year','temp'])
#print df2

cdf2 = df2.dropna()

final_df = pd.concat([pcdf,cdf2],ignore_index=True)
gdf2 =  final_df.groupby('year')
max_temps = gdf2['temp'].max()

max_temps.plot(kind='bar')
plt.show()
```

```python
#Pivot table
#Achieve same thing in diffrent way.

pivot_maxes = final_df.pivot_table('temp', aggfunc=np.max, index='year')
pivot_maxes

pivot_maxes.plot(kind='bar')
plt.ylabel('Max Temperature')
plt.show()
```
