```python

import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
```

```python
path = "/home/sejal/Documents/datascience/assignments/data/ipl_dataset.csv"
def read_csv_data_to_df(path):
    df = pd.read_csv(path)
    return df
ipl_df = read_csv_data_to_df(path)
```

```python
ipl_df.head(5)
```

```python
def plot_deliveries_by_team():
    pddf = ipl_df.groupby("batting_team")['delivery'].count().sort_values(ascending=False)
    pddf.plot(kind="bar")
    plt.show()
plot_deliveries_by_team()
```

```python
def plot_matches_by_team():
    pd_df = ipl_df.groupby("batting_team")["match_code"].nunique()
    pd_df.plot(kind="bar")
    plt.show()
plot_matches_by_team()
```

```python
def plot_innings_runs_histogram():
    #pd_df1 = ipl_df.groupby(["match_code","inning"])["runs"].sum()
    pd_df1 = ipl_df.pivot_table(columns ="inning",values = "runs", index="match_code", aggfunc="sum")   
    #print pd_df1[1]
    plt.subplot(121)
    pd_df1[1].plot(kind="hist")
    
    plt.subplot(122)
    pd_df1[2].plot(kind="hist")
    plt.show()
plot_innings_runs_histogram()
```
