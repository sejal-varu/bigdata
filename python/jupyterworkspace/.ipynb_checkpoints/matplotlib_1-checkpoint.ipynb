```python
import matplotlib.pyplot as plt

#Use https://matplotlib.org for more functions
```

```python
pointers = [10,10,10,4,5,6,2,1,1,10,7]
```

```python
plt.plot(pointers,'ro') # red color , dotted chart
#plt.plot(pointers) # default color, default chart type-LINE
plt.ylabel('Pointers')
plt.show()
```

Defualt is line graph

```python
roll_nos = [10, 4, 3,7, 1, 2]
marks = [6, 10, 5, 5, 10, 10]
plt.axis([0,15,0,12]) # AXIS TAKES LIST OF X-MIN X-MAX Y-MIN Y-MAX, GIVE HIGHER RANGES TO SEE PROPER DATA PLOTTED
#plt.plot(roll_nos, marks,'ro')
plt.xlabel('Roll numbers')
plt.ylabel('Marks')
plt.bar(roll_nos, marks) # bar() to get bar graph
#plt.barh(marks, roll_nos) # barh() to get horizontal bar graph
plt.show()
```

```python
categories = ['Travel', 'Food', 'Communications','Home']
expenses = [3000, 5000, 500, 40000]
colors = ['yellow','red','blue','green']
plt.pie(expenses, labels=categories, colors=colors, shadow=True)
plt.show()

#Dont want to show graph here, Save In a file. Comment out show() when use savefig()
#plt.savefig('/home/sejal/Documents/datascience/python/graphDocs/expenses', format='png') #first arg is path, second is filetype OR extension
```

```python
instrumments = ['Equities','MF','FD','Bonds']
investments = [800000,400000,30000,50000]
plt.pie(investments,labels=instrumments)
plt.show()
```

```python

#SUB-PLOT CONCEPT

categories = ['Travel', 'Food', 'Communications','Home']
expenses = [3000, 5000, 500, 40000]
colors = ['yellow','red','blue','green']
plt.subplot(211) #Subplot is used to show figure by plot address 211, do not use show() when assign plot.
plt.pie(expenses, labels=categories, colors=colors, shadow=True)

instrumments = ['Equities','MF','FD','Bonds']
investments = [800000,400000,30000,50000]
plt.subplot(212)
plt.pie(investments,labels=instrumments)

#plt.show()

#To save this in file
plt.savefig('/home/sejal/Documents/datascience/python/graphDocs/twoinone', format='png') #first arg is path with filename(NO extension), second is filetype OR extension
```
