```python
data.head()
X = data.iloc[:,:-1]
y = data.iloc[:,-1]

lr.fit(X,y)
# sub apne laptop ka audio band karo n confernce call pe bolo
y_pred = lr.predict(X)

mse = mean_squared_error(y,y_pred)
mae = mean_absolute_error(y,y_pred)
r2 = r2_score(y,y_pred)
print(mse,mae,r2)
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.30)

lr.fit(X_train,y_train)
y_pred1 = lr.predict(X_test)
y_pred1

mse1 = mean_squared_error(y_test,y_pred1)
mae1 = mean_absolute_error(y_test,y_pred1)
r2_1 = r2_score(y_test,y_pred1)

print(mse1,mae1,r2_1)
```

```python
pf = PolynomialFeatures(degree = 3)
lr.fit((pf.fit_transform(X_train)),y_train)
y_pred2  = lr.predict(pf.transform(X_test))

mse2 = mean_squared_error(y_test,y_pred2)
mae2 = mean_absolute_error(y_test,y_pred2)
r2_2 = r2_score(y_test,y_pred2)

print(mse2,mae2,r2_2)

#lr.coef_
```

```python
def outlier_range(data):

    data2 = data.copy()

    for col in l:
        data2[col] = data2[col].fillna(data2[col].mean())
        
        
    quantile_value1 = data2.quantile(0.02)
    quantile_value2 = data2.quantile(0.99)
    index = quantile_value1.index.values
    
    for i in range(len(l)):
        data2 = data2[(data2[index[i]] >= quantile_value1[i]) & (data2[index[i]] <= quantile_value2[i]) ]
    

    return data2
outlier_range(data)
data2 =  outlier_range(data)
data2
```

```python
#kya    hua

from scipy.stats import skew

def skewnesslog1(data2):
    
    for col in l:
        data2[col] = data2[col].fillna(data2[col].mean())

    for col in l:
       
        print(skew(data2[col]))
    print('--------')
        
    
    for i in l:
        data2[i] = np.sqrt(data2[i])

    
    for i in l:
         print(skew(data2[i]))
            

skewnesslog1(data2)
```

```python
pf = PolynomialFeatures(degree = 2 )

lr.fit((pf.fit_transform(X_train)),y_train)
y_pred2  = lr.predict(pf.transform(X_test))

mse2 = mean_squared_error(y_test,y_pred2)
mae2 = mean_absolute_error(y_test,y_pred2)
r2_2 = r2_score(y_test,y_pred2)

print(mse2,mae2,r2_2)
```
